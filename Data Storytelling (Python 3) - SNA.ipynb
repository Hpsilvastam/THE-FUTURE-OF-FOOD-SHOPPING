```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
#“The Instacart Online Grocery Shopping Dataset 2017”, Accessed from https://www.instacart.com/datasets/grocery-shopping-2017 on <09-16-2017>
```

Continuing the exploration of the data we will read the data frames created in spark, due to memmory limitations, and create the graph for visualization in Gephi.
We have 3 dataframes to explore the one connecting the user with department, aisle and product.

```python
user_aisle = pd.read_csv('Instacart Kaggle/user_aisle.csv')
user_aisle.sort_values(ascending=False, by='count').head(1000)
```

```python
pd.unique(user_aisle.aisle_id)
```

```python
#In case crashes limit by the count 
u_a = user_aisle[user_aisle['count'] > 50]
G = nx.Graph()
G.add_nodes_from(u_a['user_id'], bipartite='users')
G.add_nodes_from(u_a['aisle_id'], bipartite='aisles')
len(G.nodes())
```

```python
G.add_edges_from(zip(u_a['user_id'], u_a['aisle_id']))
len(G.edges())
```

```python
user_nodes = [n for n, d in G.nodes(data=True) if d['bipartite'] == 'users']
aisle_nodes = [n for n, d in G.nodes(data=True) if d['bipartite'] == 'aisles']
G_aisle = nx.bipartite.weighted_projected_graph(G, aisle_nodes)
len(G_aisle.nodes())
```

```python
G_aisle.nodes(data=True)
```

```python
G_aisle.edges(data=True)
base1 = 50
base2 = 500
edges = [(u, v, d) for u, v, d in G_aisle.edges(data=True) if d['weight'] >= base1 | d['weight'] <= base2]
edges
```

```python
nx.write_graphml(G_aisle, "user_aisle.graphml") 
```
